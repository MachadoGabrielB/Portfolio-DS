```python
# Algoritmo que solicita três lados de um triângulo, com esses dados o nosso programa deve informar a nomenclatura dos triângulos.
# Triângulo Equilátero: três lados iguais;
# Triângulo Isósceles: quaisquer dois lados iguais;
# Triângulo Escaleno: três lados diferentes;

list_1 =[]

valid = True
while valid == True :
  for i in range(3):

    valid2 = True
    while valid2 == True:
      lado = (input(f'Adicione o valor do lado {i+1} : '))
      if lado.isnumeric() == False :
        print('Valor inserido inválido, favor inserir novamente.')
      elif float(lado) <= 0 :
        print('Valor inválido, favor inserir novamente.')
      else:
        list_1.append(float(lado))
        valid2 = False

  list_1 = sorted(list_1, reverse= True)

  if list_1[0] < list_1[1] + list_1[2] :
    valid = False
  else :
    list_1 = []
    print('Lados não atendem a proporcionalidade necessária para fazer um triângulo.')


if list_1[0] == list_1[1] and list_1[0] == list_1[2] :
  print('O triângulo é equilátero.')
elif list_1[0] == list_1[1] or list_1[0] == list_1[2] or list_1[1] == list_1[2] :
  print('O triângulo é isósceles.')
else :
  print('O triângulo é escaleno')

```

```python
# Em uma competição de salto em distância cada atleta tem direito a cinco saltos. No final da série de saltos de cada atleta, 
# o melhor e o pior resultados são eliminados. O seu resultado fica sendo a média dos três valores restantes. 
# Você deve fazer um programa que receba o nome e as cinco distâncias alcançadas pelo atleta em seus saltos e 
# depois informe a média dos saltos conforme a descrição acima informada (retirar o melhor e o pior salto e depois calcular a média). 
# Faça uso de uma lista para armazenar os saltos. Os saltos são informados na ordem da execução, portanto não são ordenados. 
# O programa deve ser encerrado quando não for informado o nome do atleta. A saída do programa deve ser conforme o exemplo abaixo:

# Atleta: Rodrigo Curvêllo

# Primeiro Salto: 6.5 m
# Segundo Salto: 6.1 m
# Terceiro Salto: 6.2 m
# Quarto Salto: 5.4 m
# Quinto Salto: 5.3 m

# Melhor salto:  6.5 m
# Pior salto: 5.3 m
# Média dos demais saltos: 5.9 m

# Resultado final:
# Rodrigo Curvêllo: 5.9 m
```

```python




# Criando dicionário da informação

dict_1 = {
    'atleta' : 'Rodrigo Curvêllo' ,
    'saltos' : [6.5,
                6.1,
                6.2,
                5.4,
                5.3]
    }


# Adicionando erro na ausência do nome do atleta

if dict_1['atleta'] != '' :

    # Buscando valor máximo e mínimo
    max_salto = 0
    min_salto = 100

    for a in dict_1['saltos'] :
        if a > max_salto :
            max_salto = a
        if a < min_salto :
            min_salto = a

    # Removendo a valores máximo e mínimo da lista

    lista_media = dict_1['saltos']

    lista_media.remove(min_salto)
    lista_media.remove(max_salto)


    # Calculando a média

    media = sum(lista_media) / len(lista_media)

    # Mostrando o resultado.

    atleta = dict_1['atleta']

    f'O atelta {atleta} teve uma média de {media}'

else:
    print('operação inválida.')
```
